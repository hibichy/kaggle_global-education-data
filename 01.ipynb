```python
import matplotlib.pyplot as plt
import seaborn as sns

education_data = df
# Setting the aesthetic style of the plots
sns.set(style="whitegrid")

# Creating subplots for literacy rates and enrollment rates
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Literacy Rates for Youth (Male and Female)
sns.histplot(education_data['Youth_15_24_Literacy_Rate_Male'], bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('15-24歳男性の識字率')
axes[0, 0].set_xlabel('識字率 (%)')
axes[0, 0].set_ylabel('国の数')

sns.histplot(education_data['Youth_15_24_Literacy_Rate_Female'], bins=20, kde=True, color="salmon", ax=axes[0, 1])
axes[0, 1].set_title('15-24歳女性の識字率')
axes[0, 1].set_xlabel('識字率 (%)')
axes[0, 1].set_ylabel('国の数')

# Enrollment Rates for Primary and Tertiary Education
sns.histplot(education_data['Gross_Primary_Education_Enrollment'], bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('プライマリー教育の総就学率')
axes[1, 0].set_xlabel('就学率 (%)')
axes[1, 0].set_ylabel('国の数')

sns.histplot(education_data['Gross_Tertiary_Education_Enrollment'], bins=20, kde=True, color="green", ax=axes[1, 1])
axes[1, 1].set_title('高等教育の総就学率')
axes[1, 1].set_xlabel('就学率 (%)')
axes[1, 1].set_ylabel('国の数')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Visualizing gender disparities in out-of-school rates (OOSR) for different age groups

# Setting up the figure
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Pre-primary age group
sns.boxplot(data=education_data, x='OOSR_Pre0Primary_Age_Male', y='OOSR_Pre0Primary_Age_Female', ax=axes[0, 0])
axes[0, 0].set_title('Pre-primary Age Group OOSR')
axes[0, 0].set_xlabel('Male')
axes[0, 0].set_ylabel('Female')

# Primary age group
sns.boxplot(data=education_data, x='OOSR_Primary_Age_Male', y='OOSR_Primary_Age_Female', ax=axes[0, 1])
axes[0, 1].set_title('Primary Age Group OOSR')
axes[0, 1].set_xlabel('Male')
axes[0, 1].set_ylabel('Female')

# Lower secondary age group
sns.boxplot(data=education_data, x='OOSR_Lower_Secondary_Age_Male', y='OOSR_Lower_Secondary_Age_Female', ax=axes[1, 0])
axes[1, 0].set_title('Lower Secondary Age Group OOSR')
axes[1, 0].set_xlabel('Male')
axes[1, 0].set_ylabel('Female')

# Upper secondary age group
sns.boxplot(data=education_data, x='OOSR_Upper_Secondary_Age_Male', y='OOSR_Upper_Secondary_Age_Female', ax=axes[1, 1])
axes[1, 1].set_title('Upper Secondary Age Group OOSR')
axes[1, 1].set_xlabel('Male')
axes[1, 1].set_ylabel('Female')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Calculating correlations between educational and economic indicators
correlation_matrix = education_data[['Youth_15_24_Literacy_Rate_Male', 'Youth_15_24_Literacy_Rate_Female',
                                     'Gross_Primary_Education_Enrollment', 'Gross_Tertiary_Education_Enrollment',
                                     'Unemployment_Rate', 'Birth_Rate']].corr()

# Visualizing the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation between Educational and Economic Indicators')
plt.show()
```
